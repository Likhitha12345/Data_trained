```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(r'C:\Users\rjhcl\Desktop\Mushroom project\mushrooms.csv')
```

```python
df=pd.get_dummies(df)
df.head()
```

```python
df.sum().isnull()
# no null values
```

```python
X = df.drop(['class_e'] , axis=1)
y = df['class_e']
```

```python
X_train, X_test, y_test, y_train = train_test_split(X, y, test_size= 0.2)
```

```python
kmeans = KMeans(n_clusters =  3)
```

```python
kmeans.fit(X_train , y_train)
```

```python
kmean_pred = kmeans.predict(X_train)
```

```python
import pickle
with open("model","wb") as m:
    pickle.dump(kmean_pred, m)
```
