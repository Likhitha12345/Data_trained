# visualization

```python
# graph for top 10 happiest countries

labels = data["Country"][:10]
values = data["Happiness Score"][:10]

# Changing the color of bars in the plot
plt.bar(labels, values, color=['blue', 'magenta', 'pink', 'red', 'red', 'red', 'blue','red','green','green'])

plt.xticks(rotation=90)
plt.ylim(7.2, 7.6)

plt.title("Average Happiness Score Per Country")
plt.ylabel('Happiness Score (Avg)')
plt.xlabel('Country')

# Show plot
plt.tight_layout()
plt.show()

# the top happiest country is Switzerland
```
